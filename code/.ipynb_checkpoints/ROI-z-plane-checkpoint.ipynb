# Duplicate ROI over z-plane

This code will use ROI outputs to identify cell ROIs that are duplicated across z dimensions

Goal: Should we use z-plane separately? Or should we just use maximum projection? As of now, I think max projection captures everything we need and we can let the algorithms separate overlapping components based on their temporal profiles

Load in all variables

```python
plane_dirs = [os.path.join(output_path,i) for i in os.listdir(output_path) if 'plane' in i]

# get results
stats_file = []; iscell = []; f_cells = []; f_neuropils = []; spks = []; ops = []
for i in plane_dirs:
    temp_ops = np.load(os.path.join(i,'ops.npy'), allow_pickle=True).item()
    temp_ops.keys() == temp_ops.keys()  
    ops.append(temp_ops)
    stats_file.append(np.load(os.path.join(i,'stat.npy'),allow_pickle=True))
    iscell.append(np.load(os.path.join(i,'iscell.npy'),allow_pickle=True)[:, 0].astype(bool))
    f_cells.append(np.load(os.path.join(i,'F.npy'),allow_pickle=True))
    f_neuropils.append(np.load(os.path.join(i,'Fneu.npy'),allow_pickle=True))
    spks.append(np.load(os.path.join(i,'spks.npy'),allow_pickle=True))

# get image masks
im = []
for i in range(len(stats_file)):
    temp = suite2p.ROI.stats_dicts_to_3d_array(stats_file[i], Ly=ops[i]['Ly'], Lx=ops[i]['Lx'], label_id=True)
    #temp[temp == 0] = np.nan
    temp[temp > 0] = 1.0
    im.append(temp)

# get single matrix per plane
plane_mask = []
for i in range(len(im)):
    plane_mask.append(np.nanmax(im[i][iscell[i]], axis=0))
```

```python
fig, ax = plt.subplots(nrows=1,ncols=3,figsize=(10,10))
for i in range(len(plane_mask)):
    ax[i].imshow(plane_mask[i])
fig2, ax2 = plt.subplots(nrows=1,ncols=3,figsize=(10,10))
for i in range(len(plane_mask)):
    ax2[i].imshow(ops[i]['max_proj'])
```

Separate data into different planes (matrices as list)

```python
plane_rois = [im[i][iscell[i]] for i in range(len(im))]
z = len(im)
```

Flatten the arrays and concatenate across components to get a sparse (samples x components) matrix

```python
spatial = []
for planei in plane_rois:
    spatial.append(np.array([i.flatten() for i in planei]).T)
```

```python
import rois, visualization
```

Get the maximum projection (suite2p excludes data around the borders which is problematic for ROI overlay)

```python
fname = ops[0]['filelist'][0]
images = tf.memmap(fname) # lazy load
images_z = [images[zi::z] for zi in range(z)] # split the file into a list of images (planes)
max_proj = [np.max(i,axis=0).astype('float64') for i in images_z] # max projection over each plane
```

Get the intersection over union of ROIs across the zplane

```python
dims=max_proj[0].shape
templates=max_proj
spatial_union, assignments, matchings = rois.register_multisession(A=spatial, dims=dims, templates=templates)
```

```python
# Filter components by number of sessions the component could be found
n_reg = 2  # minimal number of sessions that each component has to be registered in

# Use number of non-NaNs in each row to filter out components that were not registered in enough sessions
assignments_filtered = np.array(np.nan_to_num(assignments[np.sum(~np.isnan(assignments), axis=1) >= n_reg]), dtype=int);

# Use filtered indices to select the corresponding spatial components - THIS NEEDS ADJUSTING
# The code below represents components observed on session 1 also observed on other sessions (planes)
spatial_filtered = spatial[0][:, assignments_filtered[:, 0]]
```

Check this:

I believe I have filtered out cells in plane0 that were observed in at least two planes and compared them against the total number of observed cells in plane0. 

Conclusion so far: segmented across z-plane is probably too much work for minimal change to cell count yield. We increase the number of preprocessing steps for minimal yield and it will probably take far more time to sift through the cells to confirm that our rejection criterion are adequate (temporal correlation > 0.5 + spatial union).

**Max projection is probably just better**. Lose a little yield. Maybe get some bleed through from perfectly overlapping cells, but overlapping cells with slightly diferently temporal profiles should get removed by suite2p anyway.

I do need to optimize this intersection of union code to cross register sessions regardless
